### **이미지 가져오기**

```python
#이미지 가져오기 imread를 경로를 설정하여 가져오자
image = cv2.imread('./images/castara.jpeg')
```

### **이미지 보기** 

```python
from matplotlib import pyplot as plt

#가져온 이미지를 matplolib를 사용하여 display해보자
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

이미지를 보기위한 간단한 함수를 만들어보자

```python
def imshow(title="",image = None):
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
#만든 함수를 확인해보자.
imshow("sea",image)
```

### **이미지 저장하기**

```python
#imwrite을 사용하여 저장할 이미지와 저장시 원하는 파일이름을 설정해줄수있다.
cv2.imwrite('output.jpg',image)
```

```python
#png로도 저장이 가능하다(portable network graphics)
cv2.imwrite('output.png',image)
```
