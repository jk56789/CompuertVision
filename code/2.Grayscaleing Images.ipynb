###**이미지 다운로드**

```python
import cv2 
from matplotlib import pyplot as plt
```

```python
#이미지를 다운로드하고 압축을 풀어주자
!wget https://moderncomputervision.s3.eu-west-2.amazonaws.com/images.zip
!unzip -qq images.zip
```

```python
#이미지를 보여주는 함수를 정의해보자
def imshow(title = "", image = None , size = 10):
  w,h = image.shape[0],image.shape[1] #가로,세로
  aspect_ratio = w/h
  plt.figure(figsize = (size*aspect_ratio,size))#figsize =(너비,높이)
  #plt.figure(figsize = (1,1))  #창의 크기 설정 ex)plt.figure(figsize = (9,9)) 가로9세로9인 창을 띄운다
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
#이미지를 가져오고 보여보자
image = cv2.imread('./images/castara.jpeg')
imshow("sea",image, size=6) #영어만 가능 
```

```python
image.shape #(1280, 960, 3)
```

```python
#cvtColor을 이용해 회색으로 바꾸자
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) #(가져온)이미지를 BGR에서 GRAY로바꾸자

imshow("GRAY_sea",gray_image)
```
